```python
# -----------------------------------
# Install required packages
# -----------------------------------
!pip install peft accelerate datasets transformers scikit-learn seaborn -q

# -----------------------------------
# Imports
# -----------------------------------
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification, get_cosine_schedule_with_warmup
from peft import LoraConfig, get_peft_model, TaskType
from datasets import load_dataset, Dataset
from torch.optim import AdamW
from torch.utils.data import DataLoader
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import pandas as pd
import time
from tqdm import tqdm
from accelerate import Accelerator
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns

# -----------------------------------
# Initialize Accelerator
# -----------------------------------
accelerator = Accelerator()
device = accelerator.device
print(f"Using device: {device}")

# -----------------------------------
# Load and prepare SST-2 dataset
# -----------------------------------
dataset = load_dataset("glue", "sst2")
df = pd.DataFrame(dataset['train'])

# Sample 15k for training
train_sample_df = df.groupby('label', group_keys=False).apply(
    lambda x: x.sample(n=int(15000 * len(x) / len(df)), random_state=42)
)

remaining_df = df.drop(train_sample_df.index)
test_pos = remaining_df[remaining_df.label == 1].sample(n=1500, random_state=42)
test_neg = remaining_df[remaining_df.label == 0].sample(n=1500, random_state=42)
test_df = pd.concat([test_pos, test_neg]).sample(frac=1, random_state=42).reset_index(drop=True)

train_sample, val_sample = train_test_split(
    train_sample_df, test_size=0.1, stratify=train_sample_df['label'], random_state=42
)

train_sample = Dataset.from_pandas(train_sample.reset_index(drop=True))
val_sample = Dataset.from_pandas(val_sample.reset_index(drop=True))
test_sample = Dataset.from_pandas(test_df)

# -----------------------------------
# Tokenization
# -----------------------------------
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def tokenize_batch(batch):
    return tokenizer(batch["sentence"], padding="max_length", truncation=True, max_length=128)

train_enc = train_sample.map(tokenize_batch, batched=True)
val_enc = val_sample.map(tokenize_batch, batched=True)
test_enc = test_sample.map(tokenize_batch, batched=True)

train_enc.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
val_enc.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
test_enc.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])

batch_size = 32
train_loader = DataLoader(train_enc, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_enc, batch_size=batch_size)
test_loader = DataLoader(test_enc, batch_size=batch_size)

# -----------------------------------
# Load model & Apply LoRA
# -----------------------------------
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=2)

lora_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,
    r=16,
    lora_alpha=32,
    lora_dropout=0.2,  # increased dropout to regularize
    bias="none"
)
model = get_peft_model(model, lora_config)

# Freeze non-LoRA parameters
for name, param in model.named_parameters():
    if "lora" not in name.lower():
        param.requires_grad = False

def print_trainable_parameters(model):
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Trainable params: {trainable_params} / {total_params} "
          f"({100 * trainable_params / total_params:.2f}%)")

print_trainable_parameters(model)

# -----------------------------------
# Optimizer
# -----------------------------------
learning_rate = 2e-5  # reduced LR
optimizer = AdamW(model.parameters(), lr=learning_rate, eps=1e-6 , weight_decay=0.01)



model, optimizer, train_loader, val_loader = accelerator.prepare(
    model, optimizer, train_loader, val_loader
)

# -----------------------------------
# Evaluation Function
# -----------------------------------
def evaluate(dataloader, loss_fn):
    model.eval()
    correct, total, total_loss = 0, 0, 0.0
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch['input_ids']
            attention_mask = batch['attention_mask']
            labels = batch['label']

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            preds = torch.argmax(outputs.logits, dim=1)
            loss = loss_fn(outputs.logits, labels)

            total_loss += loss.item()
            correct += (preds == labels).sum().item()
            total += labels.size(0)

    acc = correct / total * 100
    avg_loss = total_loss / len(dataloader)
    return acc, avg_loss

# -----------------------------------
# Training Loop (15 epochs fixed)
# -----------------------------------
epochs = 15
loss_fn = torch.nn.CrossEntropyLoss(label_smoothing=0.1)  # label smoothing helps generalization
loss_values, val_loss_values, val_acc_values = [], [], []
start_time = time.time()

for epoch in range(epochs):
    model.train()
    total_loss = 0.0
    epoch_start = time.time()

    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
        optimizer.zero_grad(set_to_none=True)
        outputs = model(
            input_ids=batch['input_ids'],
            attention_mask=batch['attention_mask'],
            labels=batch['label']
        )
        # Add small gradient noise to regularize
        loss = outputs.loss + 1e-5 * torch.randn(1, device=device)

        accelerator.backward(loss)
        accelerator.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / len(train_loader)
    loss_values.append(avg_loss)

    val_acc, val_loss = evaluate(val_loader, loss_fn)
    val_acc_values.append(val_acc)
    val_loss_values.append(val_loss)

    epoch_time = time.time() - epoch_start
    print(f"\nEpoch {epoch+1} | Train Loss: {avg_loss:.4f} | Val Loss: {val_loss:.4f} | Val Acc: {val_acc:.2f}% | Time: {epoch_time/60:.2f} min")

    if torch.cuda.is_available():
        allocated = torch.cuda.memory_allocated() / 1024**2
        max_allocated = torch.cuda.max_memory_allocated() / 1024**2
        print(f"GPU Memory Usage: {allocated:.2f} MB (Max: {max_allocated:.2f} MB)")

elapsed_time = time.time() - start_time
print(f"\nTotal Training Time: {elapsed_time/60:.2f} min")

# -----------------------------------
# Plot Training & Validation Metrics
# -----------------------------------
plt.figure(figsize=(8,5))
plt.plot(range(1, epochs+1), loss_values, marker='o', label='Training Loss')
plt.plot(range(1, epochs+1), val_loss_values, marker='s', label='Validation Loss')
plt.title('Training & Validation Loss vs Epoch')
plt.xlabel('Epoch'); plt.ylabel('Loss')
plt.legend(); plt.grid(True); plt.show()

# -----------------------------------
# Final Evaluation on Test Set + Confusion Matrix
# -----------------------------------
model.eval()
all_preds, all_labels = [], []

with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)

        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        preds = torch.argmax(outputs.logits, dim=1)

        all_preds.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())

test_acc = (torch.tensor(all_preds) == torch.tensor(all_labels)).sum().item() / len(all_labels) * 100
print(f"\nFinal Test Accuracy: {test_acc:.2f}%")

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Negative', 'Positive'], yticklabels=['Negative', 'Positive'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix on Test Set')
plt.show()

print("\nClassification Report:")
print(classification_report(all_labels, all_preds, target_names=['Negative', 'Positive']))

```
