```python
# -----------------------------------
# Install required packages
# -----------------------------------
!pip install peft accelerate bitsandbytes datasets transformers scikit-learn matplotlib seaborn -q

# -----------------------------------
# Imports
# -----------------------------------
import torch
from transformers import AutoTokenizer, AutoModelForSequenceClassification
from peft import LoraConfig, get_peft_model, TaskType
from datasets import load_dataset, Dataset
from torch.optim import AdamW
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import pandas as pd
import time
from tqdm import tqdm
from sklearn.model_selection import train_test_split
from accelerate import Accelerator
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay, classification_report
import seaborn as sns

# -----------------------------------
# Initialize accelerator
# -----------------------------------
accelerator = Accelerator()
device = accelerator.device
print(f"Using device: {device}")

# -----------------------------------
# Load and prepare dataset (USE OFFICIAL GLUE VALIDATION)
# -----------------------------------
dataset = load_dataset("glue", "sst2")
train_df = pd.DataFrame(dataset['train'])
val_df = pd.DataFrame(dataset['validation'])   # ✅ official GLUE validation (872 samples)

# Sample 15k for training (optional subsampling to reduce training size)
train_sample_df = train_df.groupby('label', group_keys=False).apply(
    lambda x: x.sample(n=int(15000 * len(x) / len(train_df)), random_state=42)
)

# Create a balanced test set from remaining training data (custom)
remaining_df = train_df.drop(train_sample_df.index)
test_pos = remaining_df[remaining_df.label == 1].sample(n=1500, random_state=42)
test_neg = remaining_df[remaining_df.label == 0].sample(n=1500, random_state=42)
test_df = pd.concat([test_pos, test_neg]).sample(frac=1, random_state=42).reset_index(drop=True)

# Convert to Dataset objects
train_sample = Dataset.from_pandas(train_sample_df.reset_index(drop=True))
val_sample = Dataset.from_pandas(val_df.reset_index(drop=True))      # ✅ GLUE validation
test_sample = Dataset.from_pandas(test_df)

# -----------------------------------
# Tokenizer
# -----------------------------------
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

def tokenize_batch(batch):
    return tokenizer(batch["sentence"], padding="max_length", truncation=True, max_length=128)

train_enc = train_sample.map(tokenize_batch, batched=True)
val_enc = val_sample.map(tokenize_batch, batched=True)
test_enc = test_sample.map(tokenize_batch, batched=True)

train_enc.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
val_enc.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])
test_enc.set_format(type="torch", columns=["input_ids", "attention_mask", "label"])

batch_size = 32
train_loader = DataLoader(train_enc, batch_size=batch_size, shuffle=True, pin_memory=True)
val_loader = DataLoader(val_enc, batch_size=batch_size, shuffle=False, pin_memory=True)
test_loader = DataLoader(test_enc, batch_size=batch_size, shuffle=False, pin_memory=True)

# -----------------------------------
# Load model in 4-bit (QLoRA)
# -----------------------------------
model = AutoModelForSequenceClassification.from_pretrained(
    "bert-base-uncased",
    load_in_4bit=True,
    device_map="auto",
    bnb_4bit_use_double_quant=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=torch.float32,
)

# Apply LoRA adapters
lora_config = LoraConfig(
    task_type=TaskType.SEQ_CLS,
    r=16,
    lora_alpha=32,
    lora_dropout=0.1,
    bias="none"
)
model = get_peft_model(model, lora_config)
model.to(device)

# Freeze non-LoRA parameters
for name, param in model.named_parameters():
    if "lora" not in name.lower():
        param.requires_grad = False

def print_trainable_parameters(model):
    total_params = sum(p.numel() for p in model.parameters())
    trainable_params = sum(p.numel() for p in model.parameters() if p.requires_grad)
    print(f"Trainable params: {trainable_params} / {total_params} "
          f"({100 * trainable_params / total_params:.2f}%)")

print_trainable_parameters(model)

# -----------------------------------
# Optimizer & Accelerator prep
# -----------------------------------
optimizer = AdamW(model.parameters(), lr=2e-5, eps=1e-8, weight_decay=0.01)
model, optimizer, train_loader, val_loader = accelerator.prepare(
    model, optimizer, train_loader, val_loader
)

# -----------------------------------
# Evaluation function
# -----------------------------------
def evaluate(dataloader, calc_loss=False):
    model.eval()
    correct, total = 0, 0
    total_loss = 0.0
    criterion = torch.nn.CrossEntropyLoss()
    with torch.no_grad():
        for batch in dataloader:
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['label'].to(device)

            outputs = model(input_ids=input_ids, attention_mask=attention_mask)
            preds = torch.argmax(outputs.logits, dim=1)
            correct += (preds == labels).sum().item()
            total += labels.size(0)

            if calc_loss:
                total_loss += criterion(outputs.logits, labels).item()

    acc = correct / total * 100
    avg_loss = total_loss / len(dataloader) if calc_loss else None
    return acc, avg_loss

# -----------------------------------
# Training Loop
# -----------------------------------
epochs = 15
loss_values = []
val_losses = []
val_accuracies = []
start_time = time.time()

for epoch in range(epochs):
    model.train()
    total_loss = 0.0

    for batch in tqdm(train_loader, desc=f"Epoch {epoch+1}"):
        optimizer.zero_grad(set_to_none=True)
        outputs = model(
            input_ids=batch['input_ids'],
            attention_mask=batch['attention_mask'],
            labels=batch['label']
        )
        loss = outputs.loss

        if torch.isnan(loss) or torch.isinf(loss):
            print("NaN detected, skipping batch")
            continue

        accelerator.backward(loss)
        accelerator.clip_grad_norm_(model.parameters(), max_norm=1.0)
        optimizer.step()
        total_loss += loss.item()

    avg_loss = total_loss / max(1, len(train_loader))
    loss_values.append(avg_loss)

    # Evaluate validation accuracy and loss after each epoch
    val_acc, val_loss = evaluate(val_loader, calc_loss=True)
    val_accuracies.append(val_acc)
    val_losses.append(val_loss)

    print(f"\nEpoch {epoch+1} | Train Loss: {avg_loss:.6f} | Val Loss: {val_loss:.6f} | Val Accuracy: {val_acc:.2f}%")

    if torch.cuda.is_available():
        allocated = torch.cuda.memory_allocated() / 1024**2
        max_allocated = torch.cuda.max_memory_allocated() / 1024**2
        print(f"GPU Memory Usage: {allocated:.2f} MB (Max: {max_allocated:.2f} MB)")

elapsed_time = time.time() - start_time
print(f"\nTotal Training Time: {elapsed_time/60:.2f} min")

# -----------------------------------
# Plot Loss curves (Train + Val)
# -----------------------------------
plt.figure(figsize=(8,6))
plt.plot(range(1, epochs+1), loss_values, marker='o', label='Training Loss', color='b')
plt.plot(range(1, epochs+1), val_losses, marker='s', label='Validation Loss', color='r')
plt.title('Training vs Validation Loss (QLoRA, FP32)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()

# -----------------------------------
# Final Test Evaluation + Confusion Matrix
# -----------------------------------
model.eval()
all_labels, all_preds = [], []
with torch.no_grad():
    for batch in test_loader:
        input_ids = batch['input_ids'].to(device)
        attention_mask = batch['attention_mask'].to(device)
        labels = batch['label'].to(device)
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        preds = torch.argmax(outputs.logits, dim=1)

        all_labels.extend(labels.cpu().numpy())
        all_preds.extend(preds.cpu().numpy())

test_acc, _ = evaluate(test_loader)
print(f"\nFinal Test Accuracy: {test_acc:.2f}%")

# Confusion Matrix
cm = confusion_matrix(all_labels, all_preds)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Negative", "Positive"])

plt.figure(figsize=(6,6))
disp.plot(cmap="Blues", values_format="d")
plt.title("Confusion Matrix (Test Set)")
plt.grid(False)
plt.tight_layout()
plt.show()

if torch.cuda.is_available():
    print(f"Final GPU Memory Usage: {torch.cuda.memory_allocated() / 1024**2:.2f} MB")

# -----------------------------------
# Classification Report (Test Set)
# -----------------------------------
print("\nClassification Report (Test Set):")
report = classification_report(all_labels, all_preds, target_names=["Negative", "Positive"], digits=4)
print(report)
```
